```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time
```

```python
headers={'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/54.0.2840.90 Safari/537.36'}
```

```python
front_link=[]
for i in range(1,45):
    front_link.append(f"https://www.flipkart.com/search?q=+tv&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}")
    
```

```python
def Merge(dict1, dict2):
    return(dict2.update(dict1))
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_columns', None)
```

```python
a="https://www.flipkart.com"


all_dict=[]
new_list=["Name","Discounted_Price","Price","Review","%Off"]
for i in range(1,44):
    j=f"https://www.flipkart.com/search?q=+tv&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}"
    r1=requests.get(j,headers=headers)
    print(r1)
    time.sleep(1)
    soup=BeautifulSoup(r1.content,"html.parser")
    links=soup.find_all("a",class_="_1fQZEK")
    links_1=[]
    for i in links:
        mid_link=i.get("href")
        new_link=a+mid_link
        links_1.append(new_link)
    
    for i in links_1:
        r2=requests.get(i,headers=headers)
        print(r2)
        time.sleep(1)
        new_val=[]
        new_dic={}
        soup_new1=BeautifulSoup(r2.content,"html.parser")
        name=soup_new1.find("span",class_="B_NuCI")
        discounted_price=soup_new1.find("div",class_="_30jeq3 _16Jk6d")
        price=soup_new1.find("div",class_="_3I9_wc _2p6lqe")
        review=soup_new1.find("div",class_="_3LWZlK")
        off=soup_new1.find("div",class_="_3Ay6Sb _31Dcoz")
        new_val.append(name.text if name else 'N/A')
        new_val.append(discounted_price.text if discounted_price else 'N/A')
        new_val.append(price.text if price else 'N/A')
        new_val.append(review.text if review else 'N/A')
        new_val.append(off.text if off else 'N/A')
        new_dic=dict(zip(new_list,new_val))


        all_nam=soup_new1.find_all("td",class_="_1hKmbr col col-3-12")
        all_val=soup_new1.find_all("li",class_="_21lJbe")
        val=[]
        nam=[]
        dic={}

        for k in all_nam:
            nam.append(k.text)
        for l in all_val:
            val.append(l.text)
        dic=dict(zip(nam,val))
        Merge(new_dic, dic)
        print(len(dic))
        all_dict.append(dic)
```
